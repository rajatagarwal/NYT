# Fetching data from API and storing in csv format

Below function uses NYT Archive API to gather data for years 2014, 2015 and 2016. It gathers data for every 3 months. I did preprocess of data and feature selections for creating 3 different csv files for Sentimental Analysis, Top Trends Analysis and Document Trends Analysis.

```python
import json
import requests
import pandas as pd

sentimental_outfile = open("sentimental_data.csv", "w")
most_famous_outfile = open("most_famous_data.csv", "w")
doc_trends_outfile = open("doc_trends_data.csv", "w")

sentimental_df = pd.DataFrame(columns = ("id", "section", "snippet", "pub_date", "word_count"))
most_famous_df = pd.DataFrame(columns = ("id", "section", "k_name", "k_value", "pub_date"))
doc_trends_df = pd.DataFrame(columns = ("id", "section", "doc_type", "pub_date", "word_count"))

# Dump data for Sentimental Analysis
def sentimental_analysis_data_dump(response):
	global sentimental_df 
	sentimental_df = sentimental_df.append({
		'id': response["_id"],
		'section': response["section_name"],
		'snippet': response["snippet"],
		'pub_date': response["pub_date"],
		'word_count': response["word_count"]
		}, ignore_index = True)

# Dump data for Top Trends Analysis
def most_famous_data_dump(response):
	global most_famous_df
	keywords = len(response["keywords"])
    # We will check if keywords are in string format or empty.
	if(isinstance(response["keywords"], list)):
		for keyword in range(keywords): 
			if("name" in response["keywords"][keyword] and "value" in response["keywords"][keyword]):
				most_famous_df = most_famous_df.append({
					'id': response["_id"],
					'section': response["section_name"],
					'k_name': response["keywords"][keyword]["name"],
					'k_value': response["keywords"][keyword]["value"],
					'pub_date': response["pub_date"]
					}, ignore_index = True)
	elif not keywords: # Empty list
		most_famous_df = most_famous_df.append({
			'id': response["_id"],
			'section': response["section_name"],
			'k_name': NULL,
			'k_value': NULL,
			'pub_date': response["pub_date"]
			}, ignore_index = True)
	else: # String format
		most_famous_df = most_famous_df.append({
			'id': response["_id"],
			'section': response["section_name"],
			'k_name': response["keywords"]["name"],
			'k_value': response["keywords"]["value"],
			'pub_date': response["pub_date"]
			}, ignore_index = True)

# Dump data for Document trends
def doc_trends_data_dump(response):
	global doc_trends_df 
	doc_trends_df = doc_trends_df.append({
		'id': response["_id"],
		'section': response["section_name"],
		'doc_type': response["document_type"],
		'pub_date': response["pub_date"],
		'word_count': response["word_count"]
		}, ignore_index = True)


def process_response(response):
	global sentimental_df
	for item in range(len(response["response"]["docs"])):
		if(response["response"]["docs"][item]["section_name"] == "U.S." or response["response"]["docs"][item]["section_name"] == "World"):
			sentimental_analysis_data_dump(response["response"]["docs"][item])
			most_famous_data_dump(response["response"]["docs"][item])
			doc_trends_data_dump(response["response"]["docs"][item])
		
	sentimental_df.to_csv(sentimental_outfile, sep = "\t", encoding = "utf-8")
	most_famous_df.to_csv(most_famous_outfile, sep = "\t", encoding = "utf-8")
	doc_trends_df.to_csv(doc_trends_outfile, sep = "\t", encoding = "UTF-8")

def api_get_data():
	url = "https://api.nytimes.com/svc/archive/v1"
	headers = {'api-key': '3da9edc2fdff44eaa7f7eff111b9590a'}
	years = [2014, 2015, 2016]
	months = [3, 6, 9, 12]      # Each quarter data only, due to immense amount of data.
	for year in years:
		for month in months:
			print("Processing Month:", month, "of Year:", year)
			response = requests.get(url + "/" + str(year) + "/" + str(month) + ".json", headers = headers).json()
			process_response(response)

api_get_data()
```

# Sentimental Analysis 

## A) Computing Sentimental Score

Process:
* Gather all the news snippet for a month
* Cleaning data
* Combine all snippets into a string.
* Break this string to API calling limit (80000 char max.)
* Calling API and getting score. Aggregate this score for each news snippet for the month which we broke in last step.

```python
import requests
import pandas as pd

pos_score = 0
neg_score = 0
neutral_score = 0

sentimental_score_df = pd.DataFrame(columns = ("pos_US", "neg_US", "pos_World", "neg_World"))
sentimental_score_outfile = open("sentimental_score.csv", "w") 

# API calling and getting sentiment score for each document passed.
def do_Sentimental_analysis(snippet):
	global pos_score
	global neutral_score
	global neg_score
	url = "http://text-processing.com/api/sentiment/"
	response = requests.post(url, data = {"text":snippet}).json()
	pos_score = pos_score + float(response["probability"]["pos"])
	neg_score = neg_score + float(response["probability"]["neg"])
	neutral_score = neutral_score + float(response["probability"]["neutral"])

# Preprocessing documents for the World.
def process_for_world(year_month_list, data, pos_US, neg_US):
	global pos_score
	global neutral_score
	global neg_score
	global sentimental_score_df
	for year_month in year_month_list:
		pos_score = 0
		neutral_score = 0
		neg_score = 0
        # If section is World and pub_date has current year_month value
		section_year_month_data =  data[ (data["section"] == "World") & data['pub_date'].str.contains(year_month) ]
		sectional_quaterly_snippet = ""
        # Get a string of all appended snippets for the selected criterion
		for row in section_year_month_data.iterrows():
		# this is a tuple, hence first we give place then the value - snippet
			sectional_quaterly_snippet = sectional_quaterly_snippet + str(row[1]["snippet"])
			num_of_char = len(sectional_quaterly_snippet)//50
			list_sectional_snippet = list(map(''.join, zip(*[iter(sectional_quaterly_snippet)]* num_of_char)))
		for snippet in list_sectional_snippet:
			do_Sentimental_analysis(snippet)
	
		sentimental_score_df = sentimental_score_df.append({
			'pos_US': pos_US,
			'neg_US': neg_US,
			'pos_World': (pos_score / len(list_sectional_snippet)),
			'neg_World': (neg_score / len(list_sectional_snippet)),
			}, ignore_index = True)

# Preprocessing documents for the US.
def process_for_us(year_month_list, data):
	global pos_score
	global neutral_score
	global neg_score
	global sentimental_score_df
	for year_month in year_month_list:
		pos_score = 0
		neutral_score = 0
		neg_score = 0
        # If section is US and pub_date has current year_month value
		section_year_month_data =  data[ (data["section"] == "U.S.") & data['pub_date'].str.contains(year_month) ]
		sectional_quaterly_snippet = ""
        # Get a string of all appended snippets for the selected criterion
		for row in section_year_month_data.iterrows():
		# this is a tuple, hence first we give place then the value - snippet
			sectional_quaterly_snippet = sectional_quaterly_snippet + str(row[1]["snippet"])
			num_of_char = len(sectional_quaterly_snippet)//50
			list_sectional_snippet = list(map(''.join, zip(*[iter(sectional_quaterly_snippet)]* num_of_char)))
		for snippet in list_sectional_snippet:
			do_Sentimental_analysis(snippet)
	process_for_world(year_month_list, data, pos_score / len(list_sectional_snippet), neg_score / len(list_sectional_snippet))

def preprocess_sentimental_data():
	global pos_score
	global neutral_score
	global neg_score
	data = pd.read_csv("sentimental_data.csv", sep = "\t")
	year_month_list = ["2014-03", "2014-06", "2014-09", "2014-12", "2015-03", "2015-06", "2015-09", "2015-12", "2016-03", "2016-06", "2016-09", "2016-12"]
	print("Process Start.")
	process_for_us(year_month_list, data)
	print("Process Ends.")
	sentimental_score_df.to_csv(sentimental_score_outfile, sep = "\t", encoding = "utf-8")

preprocess_sentimental_data()
```

## Most famous entities

It gives following information:
* Top 10 US location discussed in NYT in past 3 years.
* Top 10 World location discussed in NYT in past 3 years.
* Top 10 US Person discussed in NYT in past 3 years.
* Top 10 World Person discussed in NYT in past 3 years.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import heapq

# Generate dataframe and display in tabular format for top-10 entities
def generate_dataframe(top_ten_list, data):
	values = []
	for item in top_ten_list:
		values.append(data.iloc[item]["k_value"])
	rank = range(1,11)
	
	top_ten_df = pd.DataFrame({
		'Rank': rank,
		'Entity': values
		})
	top_ten_df["Rank"] = top_ten_df["Rank"].apply(np.int64)
	display(HTML(top_ten_df.to_html(index = False)))

# Generate list of top 10 occurrences in the keywords
def generate_list(data):
    # Find unique values, index and count from the sub-dataframe.
	unique, index, count = np.unique(data["k_value"], return_index = True, return_counts = True)
	print("Total unique entries:", len(unique))
	data_dict = dict(zip(index, count))
    # find top 10 values in the dictionary.
	top_ten = heapq.nlargest(10, data_dict, key=data_dict.get)
	generate_dataframe(top_ten, data)

def preprocess_most_things():
	data = pd.read_csv("most_famous_data.csv", sep = "\t")
	print("===== Top 10 Locations in the US section in 3 years.=====")
	generate_list(data[ (data["section"] == "U.S.") & (data["k_name"] == "glocations") ])
	print("=====Top 10 Locations in the World in 3 years.=====")
	top_ten_location_W = generate_list(data[ (data["section"] == "World") & (data["k_name"] == "glocations") ])
	print("=====Top 10 Person in the US in 3 years.=====")
	top_ten_person_US = generate_list(data[ (data["section"] == "U.S.")  & (data["k_name"] == "persons") ])
	print("=====Top 10 Person in the World in 3 years.=====")
	top_ten_person_W = generate_list(data[ (data["section"] == "World") & (data["k_name"] == "persons") ])
	
preprocess_most_things()
```

## Document Trends

This is a quantitative analysis of NYT data. It shows following:

* Type of news document NYT uses most. 
* Average words in the article and blogs

NYT uses mostly articles which have more than 70% of share of total news. An average article around 800 words. NYT uses blogs and multimedia contents almost equally, with blog is little ahead with 15.4% and multimedia with 13%. Each NYT blogs contains around 500 words.

```python
import pandas as pd
import matplotlib.pyplot as plt

# draw pie chart for proportion of share among articles, blogs and multimedia
def draw_pie_chart(data):
	count_articles = len(data[ (data["doc_type"] == "article")]["doc_type"].tolist())
	count_blogposts = len(data[ (data["doc_type"] == "blogpost")]["doc_type"].tolist())
	count_multimedia = len(data[ (data["doc_type"] == "multimedia")]["doc_type"].tolist())
	labels = ["Articles", "Blogposts", "Multimedia"]
	counts = [count_articles, count_blogposts, count_multimedia]
	plt.figure(figsize=(5, 4))
	plt.title("Total share of news among Articles, Blogpost and Multimedia in past 3 years.")
	p = plt.pie(counts, labels=labels, colors=["red","blue","grey"], autopct="%.1f%%")
	plt.show()

# draw bar chart for average words in article and blog
def draw_bar_chart(score):
	y_pos = [0,1]
	plt.figure(figsize=(5,3))
	names = ["Article", "Blogpost"]
	# set up the bar chart
	plt.barh(y_pos, score, align='center', height = 0.2)
	plt.yticks(y_pos, names)
	plt.xlabel("Average Words")
	plt.title("Average Words Per Article and Blogpost")
	plt.show()

# Calculate average word count
def calculate_avg_word_count(data):
	word_sum = 0
    # Get word_count from each row.
	for row in data.iterrows():
		word_sum = word_sum + int(row[1]["word_count"])
	return (word_sum / len(data["word_count"].tolist()))

def average_word_count(data):
	article_data = data[ (data["doc_type"] == "article") ]
	blogpost_data = data[ (data["doc_type"] == "blogpost") ]
	multimedia_data = data[ (data["doc_type"] == "multimedia") ]
	draw_bar_chart([round(calculate_avg_word_count(article_data),2), round(calculate_avg_word_count(blogpost_data),2)])


def initialize_doc_trend_analysis():
	doc_trends_data = pd.read_csv("doc_trends_data.csv", sep="\t")
	draw_pie_chart(doc_trends_data)
	average_word_count(doc_trends_data)

initialize_doc_trend_analysis()
```
